```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../logs/a.tsv', sep='\t')
df
```

```python
df.median(axis=1)
```

```python
data = np.transpose(df.to_numpy())
```

```python
data?
```

```python
print(data)
```

```python
plt.boxplot(data, notch=True, sym='+', patch_artist=False,vert=True)
plt.xlabel('Generation index')
plt.ylabel('Score')

plt.xticks(rotation=90)

axes = plt.gca()
#axes.set_xlim([xmin,xmax])
axes.set_ylim([50000,200000])

maxPerGen = np.amax(data, axis=0)
plt.plot(range(1,1+len(maxPerGen)),maxPerGen, linestyle='--', color='r', linewidth=1, marker='.')

plt.show()
```

```python
plt.violinplot(data)
plt.xlabel('Generation index')
plt.ylabel('Score')

plt.xticks(rotation=90)

axes = plt.gca()
#axes.set_xlim([xmin,xmax])
#axes.set_ylim([50000,200000])

maxPerGen = np.amax(data, axis=0)
plt.plot(range(1,1+len(maxPerGen)),maxPerGen, linestyle='--', color='r', linewidth=1, marker='.')

plt.show()
```

```python
plt.rc('figure', figsize=(10, 3))

fig, ax = plt.subplots(1, 1, 
                         #sharey=True,
                         #sharex = 'col'
                         constrained_layout=True
                        )

parts = ax.violinplot(data, vert=True, showmedians=True, showmeans=False, showextrema=False, widths=1, bw_method='silverman')

for pc in parts['bodies']:
    pc.set_facecolor('#D43F3A')
    #pc.set_edgecolor('black')
    pc.set_alpha(.7)
```
