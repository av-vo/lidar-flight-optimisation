```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfs = pd.read_csv('../logs/performance-peel.csv')
dfl = pd.read_csv('../logs/performance-peel-large.csv')
```

```python
plt.rc('figure', figsize=(10, 4))

fig, ax = plt.subplots(1, 2, 
                         #sharey=True,
                         #sharex = 'col'
                         constrained_layout=True
                        )
for i in range(2):
    
    if i==0:
        df = dfs
    elif i==1:
        df = dfl
    
    num_cores=1
    ax[i].plot(df[df['num_cores']==num_cores]['num_executors'], 
            df[df['num_cores']==num_cores]['runtime'],
            linestyle='solid',
            color='black',
            label=str(num_cores) + ' core'
           )

    num_cores*=2
    ax[i].plot(df[df['num_cores']==num_cores]['num_executors'], 
            df[df['num_cores']==num_cores]['runtime'],
            linestyle='dashdot',
            color='black',
            label=str(num_cores) + ' cores'
           )

    num_cores*=2
    ax[i].plot(df[df['num_cores']==num_cores]['num_executors'], 
            df[df['num_cores']==num_cores]['runtime'],
            linestyle='dashed',
            color='black',
            label=str(num_cores) + ' cores'
           )

    num_cores*=2
    ax[i].plot(df[df['num_cores']==num_cores]['num_executors'], 
            df[df['num_cores']==num_cores]['runtime'],
            linestyle='dotted',
            color='black',
            label=str(num_cores) + ' cores'
           )


plt.plot([], [], ' ', label='per executor')
    
#ax.set_title('Computational performance of the distributed evaluator')
for i in range(2):
    
    if i==0:
        ax[i].set_title('Small Dataset')
    else:
        ax[i].set_title('Large Dataset')
    
    ax[i].set_ylim(0, None)
    
    ax[i].set_xlabel('Number of executors')
    ax[i].set_ylabel('Runtime (sec)')

    ax[i].set_xticks([1, 2, 4, 8, 16])
    ax[i].grid(color='black', ls = '--', lw = 0.2)

    #ax.set_yscale("log")
    #ax.grid()
    ax[i].legend()

plt.savefig('peel.svg')
```
