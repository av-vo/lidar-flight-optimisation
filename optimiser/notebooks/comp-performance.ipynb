```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../logs/performance.csv')
```

```python
df
```

```python
plt.rc('figure', figsize=(6, 4))

fig, ax = plt.subplots(1, 1, 
                         #sharey=True,
                         #sharex = 'col'
                         constrained_layout=True
                        )

num_cores=1
ax.plot(df[df['num_cores']==num_cores]['num_executors'], 
        df[df['num_cores']==num_cores]['runtime'],
        linestyle='solid',
        color='black',
        label=str(num_cores) + ' core'
       )

num_cores*=2
ax.plot(df[df['num_cores']==num_cores]['num_executors'], 
        df[df['num_cores']==num_cores]['runtime'],
        linestyle='dashdot',
        color='black',
        label=str(num_cores) + ' cores'
       )

num_cores*=2
ax.plot(df[df['num_cores']==num_cores]['num_executors'], 
        df[df['num_cores']==num_cores]['runtime'],
        linestyle='dashed',
        color='black',
        label=str(num_cores) + ' cores'
       )

num_cores*=2
ax.plot(df[df['num_cores']==num_cores]['num_executors'], 
        df[df['num_cores']==num_cores]['runtime'],
        linestyle='dotted',
        color='black',
        label=str(num_cores) + ' cores'
       )


plt.plot([], [], ' ', label='per executor')
    
ax.set_title('Computational performance of the distributed evaluator')
ax.set_xlabel('Number of executors')
ax.set_ylabel('Runtime (sec)')

ax.set_yscale("log")
#ax.grid()
ax.legend()

plt.savefig('a.svg')
```
