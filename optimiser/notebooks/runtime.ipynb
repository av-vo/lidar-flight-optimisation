```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('runtime.csv')
```

```python
df
```

```python
plt.rc('figure', figsize=(5, 4))

fig, ax = plt.subplots(1, 1, 
                         #sharey=True,
                         #sharex = 'col'
                         constrained_layout=True
                        )

ls = ['solid','dashdot','dashed','dotted']

num_cores = 1
for i in range(0, 4):
    ax.plot(df[df['num_cores']==num_cores]['num_executors'], 
            df[df['num_cores']==num_cores]['runtime'],
            linestyle=ls[i],
            color='black',
            label=str(num_cores) + ' core'
           )
    num_cores*=2


plt.plot([], [], ' ', label='per executor')
    
#ax.set_title('Computational performance of the distributed evaluator')
ax.set_xlabel('Number of executors')
ax.set_ylabel('Runtime (sec)')

ax.set_xticks([1, 2, 4, 8, 16])

ax.grid(color='black', ls = '--', lw = 0.2)

#ax.set_yscale("log")
#ax.grid()
ax.legend()

plt.savefig('peel-large.svg')
```
