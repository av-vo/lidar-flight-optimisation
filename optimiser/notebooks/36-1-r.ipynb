```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../logs/36-1-r.tsv', sep='\t')
data = np.transpose(df.to_numpy())
```

```python
#df.to_csv('../logs/16-2-reverse.tsv', sep='\t')
```

```python
plt.rc('figure', figsize=(7, 3))

fig, ax = plt.subplots(1, 1, 
                         #sharey=True,
                         #sharex = 'col'
                         #constrained_layout=True
                        )

parts = ax.violinplot(data, vert=True, showmedians=True, showmeans=False, showextrema=False, widths=.75, bw_method='silverman')

for pc in parts['bodies']:
    pc.set_facecolor('#D43F3A')
    #pc.set_edgecolor('black')
    pc.set_alpha(.6)
    
maxPerGen = np.median(data, axis=0)
plt.plot(range(1,1+len(maxPerGen)),maxPerGen, linestyle='--', color='r', linewidth=.5, marker='.')
```

```python
d = (data - 14000000)/14000000*100
```

```python
plt.rc('figure', figsize=(7, 3))

fig, ax = plt.subplots(1, 1, 
                         #sharey=True,
                         #sharex = 'col'
                         #constrained_layout=True
                        )

parts = ax.violinplot(d, vert=True, showmedians=True, showmeans=False, showextrema=False, widths=.75, bw_method='silverman')
```

```python
plt.rc('figure', figsize=(14, 5))

fig, ax = plt.subplots(1, 1, 
                         #sharey=True,
                         #sharex = 'col'
                         #constrained_layout=True
                        )

sns.set_theme(style="whitegrid")
sns.violinplot(data=data, linewidth=1, gridsize=18, cut=1, scale='count', bw='scott', width=1, inner='point', split=False, dodge=True, orient='v', color='#eeeeee', palette=None, saturation=0.75, ax=ax)


median = np.median(data, axis=0)
plt.plot(range(0,len(median)), median, linestyle='dashed', color='r', linewidth=1, marker='.')
```
